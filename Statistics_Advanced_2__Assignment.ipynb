# ‚úÖ **What is the Null Hypothesis, and How Does It Differ from the Alternative Hypothesis?**

## **1Ô∏è‚É£ Null Hypothesis (H‚ÇÄ)**

The **null hypothesis** is a statement that assumes **no effect, no difference, or no change** in the population.

It represents the **default belief** that we test against.

### Example:

H‚ÇÄ: The average weight of packets = 50 g.
(There is **no difference** from the claimed value.)

---

## **2Ô∏è‚É£ Alternative Hypothesis (H‚ÇÅ or Ha)**

The **alternative hypothesis** is the statement that **contradicts** the null hypothesis.

It suggests that there **is an effect, difference, or change**.

### Example:

H‚ÇÅ: The average weight of packets ‚â† 50 g.
(There **is a difference**.)

---

# üîç **Key Differences**

| Feature   | Null Hypothesis (H‚ÇÄ)                    | Alternative Hypothesis (H‚ÇÅ / Ha)  |
| --------- | --------------------------------------- | --------------------------------- |
| Meaning   | No effect / No difference               | Some effect / A difference exists |
| Purpose   | Starting assumption                     | Claim we want to test             |
| Direction | Always includes ‚Äú=‚Äù (equal)             | Uses ‚â†, <, or >                   |
| Decision  | Either **rejected** or **not rejected** | Accepted when H‚ÇÄ is rejected      |

# ‚úÖ **Significance Level in Hypothesis Testing**

The **significance level**, denoted by **Œ± (alpha)**, is the **threshold probability** used to decide whether to reject the null hypothesis.

It represents the **maximum risk of making a Type I error** (rejecting a true null hypothesis) that we are willing to accept.

Common values: **0.05**, **0.01**, **0.10**

---

# üéØ **Role of Significance Level (Œ±)**

The significance level determines the **decision rule** in hypothesis testing.

### ‚úî If **p-value < Œ±** ‚Üí **Reject H‚ÇÄ**

There is strong evidence against the null hypothesis.

### ‚úî If **p-value ‚â• Œ±** ‚Üí **Fail to reject H‚ÇÄ**

There is not enough evidence to reject the null hypothesis.

---

# ‚≠ê **Why Œ± is Important?**

* It sets the **boundary for making decisions**.
* It controls the **probability of a false alarm** (Type I error).
* It helps maintain the **reliability of conclusions**.

---

# üìå Example

If Œ± = 0.05:

* We accept a **5% risk** of rejecting a true null hypothesis.
* If the test gives **p = 0.03**, then:
  Since 0.03 < 0.05 ‚Üí **Reject H‚ÇÄ**.

---

# üìò **Short Exam-Ready Answer**

**The significance level (Œ±) is the probability of rejecting a true null hypothesis. It is used as a threshold to compare the p-value. If the p-value is less than Œ±, we reject H‚ÇÄ; otherwise, we fail to reject it. It controls the Type I error rate and helps decide the test outcome.**

# ‚úÖ **Type I and Type II Errors**

In hypothesis testing, two types of mistakes can occur when making decisions about the null hypothesis.

---

# üî¥ **1Ô∏è‚É£ Type I Error (False Positive)**

A **Type I error** occurs when we **reject the null hypothesis (H‚ÇÄ) even though it is actually true**.

### ‚úî Meaning

You think an effect exists, but in reality, it does NOT.

### ‚úî Example

A drug is **actually ineffective**, but the test concludes it **works**.

* True situation: Drug does NOT work (H‚ÇÄ true)
* Decision: Reject H‚ÇÄ
  ‚û°Ô∏è **Type I error**

### ‚úî Symbol

The probability of Type I error = **Œ± (significance level)**

---

# üîµ **2Ô∏è‚É£ Type II Error (False Negative)**

A **Type II error** occurs when we **fail to reject the null hypothesis even though it is false**.

### ‚úî Meaning

You fail to detect a real effect or difference.

### ‚úî Example

A drug is **actually effective**, but the test concludes it **does not work**.

* True situation: Drug works (H‚ÇÅ true)
* Decision: Fail to reject H‚ÇÄ
  ‚û°Ô∏è **Type II error**

### ‚úî Symbol

The probability of Type II error = **Œ≤**

---

# üìò **Short Exam-Ready Answer**

* **Type I Error (Œ±):** Rejecting a true null hypothesis.
  *Example: Concluding a machine is faulty when it is actually working correctly.*

* **Type II Error (Œ≤):** Failing to reject a false null hypothesis.
  *Example: Concluding a machine is working fine when it is actually faulty.*

# ‚úÖ **Difference Between a Z-test and a T-test & When to Use Each**

A **Z-test** and **T-test** are both used to test hypotheses about population means, but they differ based on sample size and whether the population standard deviation is known.

---

# üîµ **1Ô∏è‚É£ Z‚ÄìTest**

### ‚úî **When is a Z-test used?**

Use a Z-test when:

* **Sample size is large** (n ‚â• 30)
* **Population standard deviation (œÉ) is known**
* Data follows a **normal distribution**

### ‚úî Example

Testing whether the average height of students differs from 160 cm when the **population variance is known**.

---

# üî¥ **2Ô∏è‚É£ T‚ÄìTest**

### ‚úî **When is a T-test used?**

Use a T-test when:

* **Sample size is small** (n < 30)
* **Population standard deviation is NOT known**
* Sample is from a **normal or approximately normal** population

### ‚úî Example

Testing the mean weight of packets using a **small sample** when œÉ is **unknown**, so we use the **sample standard deviation**.

---

# üü© **Key Differences (Table)**

| Feature           | **Z-test**                    | **T-test**                                  |
| ----------------- | ----------------------------- | ------------------------------------------- |
| Sample size       | Large (‚â• 30)                  | Small (< 30)                                |
| Population SD (œÉ) | Known                         | Unknown                                     |
| Uses              | Z-distribution                | t-distribution                              |
| Shape             | Fixed, normal                 | Depends on degrees of freedom (wider tails) |
| Best for          | Large samples, known variance | Small samples, unknown variance             |

---

# ‚≠ê **Short Exam-Ready Answer**

* **Z-test:** Used when the sample size is large (n ‚â• 30) and the population standard deviation is known.
* **T-test:** Used when the sample size is small (n < 30) and the population standard deviation is unknown.

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Generate normal distribution data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=500)  # mean=50, std=10, sample=500

# 2. Calculate 95% confidence interval for mean
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# 95% CI using t-distribution
ci_low, ci_high = stats.t.interval(0.95, df=n-1, loc=mean, scale=std/np.sqrt(n))

print("Mean:", mean)
print("95% Confidence Interval:", (ci_low, ci_high))

# 3. Plot the histogram
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simulated Normal Distribution Data (Histogram)")
plt.show()

# 4. Box Plot
plt.boxplot(data)
plt.ylabel("Values")
plt.title("Box Plot of Simulated Normal Distribution Data")
plt.show()
```

# üìò **Explanation: What Do Z-Scores Represent?**

A **Z-score tells how many standard deviations** a data point is from the mean.

* **Z = 0** ‚Üí The value is exactly at the mean
* **Z = +1** ‚Üí 1 standard deviation **above** the mean
* **Z = ‚Äì2** ‚Üí 2 standard deviations **below** the mean
* **Z > 3 or Z < ‚Äì3** ‚Üí Usually considered **outliers**

After converting to Z-scores:

* The new **mean becomes 0**
* The new **standard deviation becomes 1**

This process is called **standardization**.

---

If you want, I can also provide:

üìå Z-score interpretation chart
üìå Outlier detection using Z-scores
üìå Box plot of Z-scores.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
np.random.seed(0)
data = np.random.normal(100, 15, 500)  # mean=100, std=15, n=500

# Function to calculate Z-scores
def calculate_z_scores(dataset):
    mean = np.mean(dataset)
    std = np.std(dataset, ddof=1)
    z_scores = (dataset - mean) / std
    return z_scores

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Mean of Z-scores:", np.mean(z_scores))
print("Std of Z-scores:", np.std(z_scores, ddof=1))

# Plot histogram of Z-scores
plt.hist(z_scores, bins=30, edgecolor='black')
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Standardized Data (Z-scores)")
plt.show()
```
