Question 2: What is sampling in statistics? Explain the differences between random
and stratified sampling.
>>**Sampling (Definition):**
Sampling is the process of selecting a **small group (sample)** from a **large population** to collect data and draw conclusions about the whole population.
**Random Sampling (Definition):**
A method in which **every member of the population has an equal chance** of being selected.
**Stratified Sampling (Definition):**
A method in which the population is **divided into groups (strata)** based on certain characteristics, and samples are **randomly chosen from each group**.


Question 3: Define mean, median, and mode. Explain why these measures of central
tendency are important.
>>**Mean, Median, and Mode â€“ Definitions & Importance**

---

### **1. Mean (Average):**

The **mean** is the **sum of all values divided by the total number of values**.
[
\text{Mean} = \frac{\text{Sum of all observations}}{\text{Number of observations}}
]
**Example:** For 5, 10, 15 â†’ Mean = (5 + 10 + 15)/3 = 10

---

### **2. Median:**

The **median** is the **middle value** when all data are arranged in order.
If there is an even number of values, the median is the **average of the two middle values**.
**Example:** For 4, 6, 8 â†’ Median = 6

---

### **3. Mode:**

The **mode** is the **most frequently occurring value** in a dataset.
**Example:** For 2, 3, 3, 4, 5 â†’ Mode = 3

---

### **Importance (Why They Are Important):**

* They show the **central tendency** or **typical value** of the data.
* Help in **summarizing large datasets** with a single representative value.
* Useful for **comparison** between different datasets.
* Provide insight into the **distribution** and **pattern** of data.

---

**In short:**

> **Mean** = average,
> **Median** = middle value,
> **Mode** = most frequent value â€”
> All three help describe where the **center** of the data lies.


Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?
>>### **Skewness and Kurtosis**

---

### **1. Skewness:**

* **Definition:** Skewness measures the **asymmetry (unevenness)** of a data distribution around its mean.
* If data are perfectly symmetric (like a normal distribution), **skewness = 0**.

| **Type**                         | **Shape**                             | **Meaning**                                                                            |
| -------------------------------- | ------------------------------------- | -------------------------------------------------------------------------------------- |
| **Positive Skew (Right Skewed)** | Tail is longer on the **right** side. | Most data values are **small**, and a few **large values** pull the mean to the right. |
| **Negative Skew (Left Skewed)**  | Tail is longer on the **left** side.  | Most data values are **large**, and a few **small values** pull the mean to the left.  |

**ðŸ‘‰ Positive Skew implies:**
The distribution has a **long right tail**, the **mean > median**, and most data values are **below the average**.

---

### **2. Kurtosis:**

* **Definition:** Kurtosis measures the **â€œpeakednessâ€ or â€œflatnessâ€** of a distribution compared to a normal curve.
* It shows how **heavy or light the tails** of a distribution are.

| **Type**                         | **Shape**                    | **Meaning**    |
| -------------------------------- | ---------------------------- | -------------- |
| **Leptokurtic (High Kurtosis)**  | Very **peaked**, heavy tails | More outliers  |
| **Mesokurtic (Normal Kurtosis)** | **Normal bell shape**        | Moderate tails |
| **Platykurtic (Low Kurtosis)**   | **Flat top**, light tails    | Fewer outliers |

---

### **In short:**

* **Skewness â†’** measures **shape symmetry**
* **Kurtosis â†’** measures **tail thickness / peak sharpness**
* **Positive skew â†’** data have a **long right tail** and **mean > median**

Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
(Include your Python code and output in the code box below.)
>>


```python
# Program to compute covariance and correlation coefficient

import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate covariance matrix
cov_matrix = np.cov(list_x, list_y, bias=False)
covariance = cov_matrix[0, 1]

# Calculate correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0, 1]

# Display results
print("Covariance =", covariance)
print("Correlation Coefficient =", correlation)
```

Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
(Include your Python code and output in the code box below.)


```python
# Program to draw a boxplot and identify outliers

import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

# Identify outliers using IQR method
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
print("Outliers =", outliers)
```

Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
â— Explain how you would use covariance and correlation to explore this
relationship.
â— Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]
(Include your Python code and output in the code box below.)


```python
# Program to compute correlation between advertising spend and daily sales

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute covariance and correlation
covariance = np.cov(advertising_spend, daily_sales, bias=False)[0, 1]
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Covariance =", covariance)
print("Correlation Coefficient =", correlation)
```

Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
â— Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) youâ€™d use.
â— Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
(Include your Python code and output in the code box below.
>>ðŸ”¹ Explanation

To understand the distribution of customer satisfaction scores (1â€“10 scale), you should use both summary statistics and visualizations:

Summary Statistics:

Mean: Average satisfaction level.

Median: Middle score showing central tendency.

Mode: Most frequent score.

Standard Deviation (SD): Measures how spread out the scores are.

Small SD â†’ customers have similar opinions.

Large SD â†’ mixed opinions.

Visualizations:

Histogram: Shows how frequently each score occurs (shape of distribution).

Boxplot (optional): Helps detect outliers and symmetry of data.

These tools together reveal if most customers are satisfied (high scores) or dissatisfied (low scores).

```python
# Program to analyze and visualize customer satisfaction survey data

import matplotlib.pyplot as plt
import statistics as stats

# Given data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary statistics
mean_score = stats.mean(survey_scores)
median_score = stats.median(survey_scores)
mode_score = stats.mode(survey_scores)
std_dev = stats.stdev(survey_scores)

print("Mean =", mean_score)
print("Median =", median_score)
print("Mode =", mode_score)
print("Standard Deviation =", std_dev)

# Histogram visualization
plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title("Customer Satisfaction Scores")
plt.xlabel("Score (1â€“10)")
plt.ylabel("Frequency")
plt.show()
```
